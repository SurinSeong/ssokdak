# 데이터셋 구성

- 파일명|원본텍스트|텍스트

```python
import pandas as pd

for_metadata = pd.read_csv("/content/drive/MyDrive/ssafy/2학기/ssokdak/data/korean-single-speaker-speech-dataset/transcript.v.1.4.txt", sep="|", header=None, encoding="utf-8")
print(for_metadata.shape)

for_metadata.head(2)
```

```python
import os
import shutil

# 대상 폴더 경로 (새로 생성될 wavs 폴더)
destination_folder = "/content/drive/MyDrive/ssafy/2학기/ssokdak/data/korean-single-speaker-speech-dataset/wavs"

for num in [1, 2, 3, 4]:
    # 원본 폴더 경로
    source_folder = f"/content/drive/MyDrive/ssafy/2학기/ssokdak/data/korean-single-speaker-speech-dataset/kss/{num}"

    # 원본 폴더의 모든 파일 순회
    for filename in os.listdir(source_folder):
        # .wav 파일인 경우에만 이동
        if filename.endswith(".wav"):
            source_path = os.path.join(source_folder, filename)
            destination_path = os.path.join(destination_folder, filename)
            # 파일 이동
            shutil.move(source_path, destination_path)

    print(f"'{source_folder}'의 .wav 파일이 '{destination_folder}'로 이동되었습니다.")
```

```python
print(len(os.listdir(destination_folder)))
```

```python
# 필요한 컬럼만 선택 (0, 1, 2 컬럼)
new_metadata = for_metadata[[0, 1, 2]].copy()

# 파일명 컬럼 (0번 컬럼)에서 '폴더명/' 제거하고 확장자 유지
# 예: '1/1_0000.wav' -> '1_0000.wav'
new_metadata[0] = new_metadata[0].apply(lambda x: os.path.basename(x))
new_metadata[0] = new_metadata[0].apply(lambda x: x.replace(".wav", ""))

# 결과 확인
new_metadata.head(2)
```

```python
new_metadata.to_csv("/content/drive/MyDrive/ssafy/2학기/ssokdak/data/korean-single-speaker-speech-dataset/metadata.txt", index=False, header=False, sep="|", encoding="utf-8")
```
